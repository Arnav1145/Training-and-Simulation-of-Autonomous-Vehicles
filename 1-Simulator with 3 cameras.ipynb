```python
#taking all the center,left and right images into the respective variable
#also taking the steering angle for each image and adjusting them accrodingly by
# adding 0.2 to left images and subtracting 0.2 from right images
center = df.iloc[:,0:1].values
center_y = df.iloc[:,3:4].values

left = df.iloc[:,1:2].values
left_y = df.iloc[:,3:4].values + 0.2

right = df.iloc[:,2:3].values
right_y = df.iloc[:,3:4].values - 0.2
```

```python
#now concatinating each of the above data into one as 
#X features and y labels
X = np.concatenate((center,left,right),axis = 0)
y = np.concatenate((center_y,left_y,right_y),axis = 0)
```

```python
X.shape
```

```python
y.shape
```

```python
data = np.concatenate((X,y), axis = 1)
data = shuffle(data)
```

```python
data
```

```python
#Most of the times steering values are 0 so we will check how many
#positive, zero and negative images

pos = 0
neg = 0
zeroes = 0

for i in y:
    if (i>0):
        pos+=1
    elif (i<0):
        neg+=1
    else:
        zeroes+=1
```

```python
print(neg,pos,zeroes)
```
